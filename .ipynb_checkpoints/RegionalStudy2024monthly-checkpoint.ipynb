```python
#Import Dependencies

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
import calendar

import warnings
warnings.filterwarnings('ignore')

```

```python
#Read in CSV file
egg_data = pd.read_csv("Resources/USDA_Eggs_2024.csv")

#Create Dataframe
egg_data
```

```python
# validate the data types in the raw data
egg_data.dtypes
```

```python
# select the fields needed for the analysis
egg_data_df = egg_data[['report_date','region','type','quality_grade','package_size','environment','price_avg', 'store_count','section']]

# filter the rows needed for the analysis - shell eggs in cartons of 12 or less
egg_data_df = egg_data_df.loc[((egg_data_df['section'] == 'Shell Egg') &\
                              ((egg_data_df['package_size'] != '30') & (egg_data_df['package_size'] != '60')))]

egg_data_df
```

```python
# change the report date from a string to a date field to support proper sorting of dates for reporting
egg_data_df['report_date'] = pd.to_datetime(egg_data_df['report_date'], format='%m/%d/%Y')
egg_data_df['report_month'] = egg_data_df['report_date'].dt.month

egg_data_df
```

```python
# validate the data type change was made
egg_data_df.dtypes
```

```python
# clean up region data
# identify the data that needs to be cleaned, validate the record counts prior to cleaning
egg_data_df['region'].value_counts()
```

```python
#clean up region data
egg_data_df['region'] = egg_data_df['region'].replace({'NATIONAL': 'National', 'Northeast': 'NorthEast',
                                     'Southeast': 'SouthEast', 'Midwest': 'MidWest',
                                     'Southcentral': 'SouthCentral', 'Northwest': 'NorthWest',
                                     'Southwest': 'SouthWest'})
egg_data_df
```

```python
# validate the changes were made and the counts make sense
egg_data_df['region'].value_counts()
```

```python
# calculate the weighted average price at the regional level

# create a field to house the total cost of eggs on each row 
egg_data_df['total_price'] = egg_data_df['store_count'] * egg_data_df['price_avg']

# sum the total price and total store count numbers by report date and region
egg_data_sum_df = egg_data_df.groupby(['report_month','region'])[['store_count','total_price']].sum()

# create a new field for the average price by report date and region, round to 2 decimal points
egg_data_sum_df['reg_avg_price'] = round((egg_data_sum_df['total_price']/egg_data_sum_df['store_count']),2)

# reset the index
egg_data_sum_df.reset_index(inplace=True)
egg_data_sum_df
```

```python
# pare the table down to the fields needed for plotting

egg_data_sum_avgprice_df = egg_data_sum_df[['report_month','region','reg_avg_price']]
egg_data_sum_avgprice_df
```

```python
# use the pivot function to create a matrix of average weighted prices by region and date for plotting
egg_data_avgprice_plot = egg_data_sum_avgprice_df.pivot(index='report_month', columns='region', values='reg_avg_price')

egg_data_avgprice_plot
```

```python
# drop columns that contain a lot of missing data
egg_data_avgprice_plot.dropna(axis='columns',inplace=True)
egg_data_avgprice_plot
```

```python
# create a line graph showing a line for each region using pandas
egg_data_avgprice_plot.plot(kind='line',
                      figsize=(12,8),
                      title='Monthly Retail Egg Prices by Region in 2024',
                      xlabel='Month', 
                      ylabel='Average Price per Carton ($)')
plt.show()
```

```python
# reset the index so the regions can be plotted individually using pyplot

egg_data_avgprice_plot.reset_index(inplace=True)
```

```python
# plot the three regions to be analyzed further
x = egg_data_avgprice_plot['report_month']

y3 = egg_data_avgprice_plot['MidWest']
y4 = egg_data_avgprice_plot['NorthEast']
y9 = egg_data_avgprice_plot['National']

plt.plot(x, y3, '--', label ='MidWest')
plt.plot(x, y4, '-.', label ='NorthEast')
plt.plot(x, y9, label ='National')
#wh_lg_conv_plot.plot(kind='line',x='report date',y='price avg')

plt.legend()
plt.figure(figsize=(800,600))
plt.figure(layout='tight')
plt.show()
```

```python
# take a look at the number of stores reporting - do they drop during the low points?
egg_data_sum_storecount_df = egg_data_sum_df[['report_date','region','store_count']]
egg_data_storecount_plot = egg_data_sum_storecount_df.pivot(index='report_date', columns='region', values='store_count')
egg_data_storecount_plot.dropna(axis='columns',inplace=True)
egg_data_storecount_plot
```

```python
egg_data_storecount_plot.plot(kind='line',
                      figsize=(12,8),
                      title='Weekly Retail Egg Stores by Region in 2024',
                      xlabel='Date', 
                      ylabel='Average #Stores Reporting')
plt.show()
```
