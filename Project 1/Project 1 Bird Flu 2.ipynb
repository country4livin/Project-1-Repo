# Bird Flu Outbreaks in the United States

Analysis: The data was provided by the CDC for the Bird Flu Outbreaks in the US starting in 2022 and focused on Table Egg Layers flock sizes affected. Data was graphed nationally. Then it was sorted and graphed regionally to see how and if the regions of the US were affected. 

The Midwest had the most outbreaks across the years. This makes sense because Iowa, Ohio and Indiana are the top 3 egg producing states thus having a large population of Table Egg Layers. The West was the next highest region to be affected, with California being the predominate state for egg production. Texas had the majority of outbreaks for the Southwest, which makes sense due to the amount of Table Layers that are in Texas. 

The Midwest and the West have reported outbreaks every year since 2022. So far in the first month of 2025 there has been over 14 million birds reported, which is more than the entire year of 2023.

The next steps could include filtering for Table Egg Pullets to see how many immature chickens are being affected. It would then be worth looking into the idea behind culling the entire flock, rather than quarantining the infected birds/flock and seeing if there are genetically superior birds that can resist the flu, then use them for breeding future birds with natural immunity. It is also worth looking into the sanitation and hygiene methods used before, during and after outbreaks and see if there is room for improvement. 

```python
import pandas as pd
import numpy as np
import requests
import json
import hvplot.pandas
import time
from scipy.stats import linregress
from matplotlib import pyplot as plt
from config import geoapify_key
from pprint import pprint
```

```python
#load CSV file for US bird flu
bird_flu_pd = pd.read_csv("Resources/commercial-backyard-flocks.csv")
bird_flu_pd
```

### National Egg Layer Outbreak Data

```python
# Filter for National bird flu 'Table Egg Layer' outbreaks
egg_layers = bird_flu_pd.loc[bird_flu_pd["Flock Type"].isin(["Commercial Table Egg Layer"])]

egg_layers
```

```python
#Combine Flock Size data for same date values into a dataframe for graphing
result = {}

for index, row in egg_layers.iterrows():
#    state = row['State']
    date = row['Outbreak Date']
    value = row['Flock Size']

    if date in result:
        result[date] += value
    else:
        result[date] = value

combined_egg_layers = pd.DataFrame(result.items(), columns=['Outbreak Date', 'Total Flock Size'])
combined_egg_layers
```

```python
#Set Flock size to numeric values
combined_egg_layers['Total Flock Size'] = pd.to_numeric(combined_egg_layers['Total Flock Size'])
combined_egg_layers
```

```python
#Set index to Outbreak Date and sort by ascending
combined_egg_layers['Outbreak Date'] = pd.to_datetime(combined_egg_layers['Outbreak Date']).dt.date
combined_egg_layers.set_index(['Outbreak Date'], inplace=True)
combined_egg_layers.sort_index(inplace=True)
combined_egg_layers
```

```python
min_flock = combined_egg_layers['Total Flock Size'].min()
print(min_flock)
max_flock = combined_egg_layers['Total Flock Size'].max()
print(max_flock)
```

```python
#Graph National Table Egg Layer data
figure1 = combined_egg_layers['Total Flock Size'].plot(kind="bar", color="red", figsize=(25, 10),
                                    title="National Table Egg Layers Outbreak since 2022",
                                    xlabel="Outbreak Date",
                                    ylabel="Flock Size")
plt.xticks(ha='center', rotation=90)
plt.bar_label(figure1.containers[0])
plt.savefig('National_Outbreak.png')
plt.show()
```

```python
# Map graph of National Outbreaks with Total FLock Size as size of dot using GeoApify
#Add Lat and Lon columns for holding Latitude and Longitude coordinates from Geoapify
#egg_layers["County State"] = egg_layers["County"]+ egg_layers["State"]
egg_layers["Lat"] = ""
egg_layers["Lon"] = ""

# Define the API parameters
params = {
    "apiKey":geoapify_key,
    "format":"json"
}

# Set the base URL
base_url = "https://api.geoapify.com/v1/geocode/search"

# Loop through the egg_layers DataFrame and search coordinates for each county
for index, row in egg_layers.iterrows():

    # Get the county's name & add County so geoapify finds the correct county
    county = row["County"] + " " + "County"

    # Add the current county to the parameters
    params["text"] = f"{county}"

    # Make the API request
    response = requests.get(base_url, params=params)
    
    # Convert response to JSON
    response = response.json()

    #pprint(response)
    # Extract latitude and longitude
    egg_layers.loc[index, "Lat"] = response["results"][0]["lat"]
    egg_layers.loc[index, "Lon"] = response["results"][0]["lon"]

egg_layers
```

```python
#Use Lat and Lon to map Flock size by County using hvplot
national_map_plot = egg_layers.hvplot.points(
    "Lon",
    "Lat",
    geo = True,
    tiles = "OSM",
    frame_width = 1000,
    frame_height = 800,
    size = "Flock Size",
    scale = 0.01,
    color = "County"
)
hvplot.save(national_map_plot, "National Map.png")
national_map_plot
```
